```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

```python
# Verificar valores nulos em todas as lojas
print("\nValores nulos por coluna em cada loja:")
for i, df in enumerate([loja1, loja2, loja3, loja4], 1):
    print(f"\nLoja {i}:")
    print(df.isnull().sum())

# Verificar duplicatas
print("\nTotal de registros duplicados por loja:")
for i, df in enumerate([loja1, loja2, loja3, loja4], 1):
    print(f"Loja {i}: {df.duplicated().sum()}")
```

Média de Avaliação por Loja

```python
import pandas as pd
import numpy as np

# DataFrame para armazenar os resultados
aval = pd.DataFrame()

try:
    # Calcular métricas para cada loja
    for i, df_loja in enumerate([loja1, loja2, loja3, loja4], 1):
        if isinstance(df_loja, pd.DataFrame) and 'Avaliação da compra' in df_loja.columns:
            avaliacoes = pd.to_numeric(df_loja['Avaliação da compra'], errors='coerce')
            aval.loc[f'Loja {i}', 'Média'] = avaliacoes.mean()
        else:
            print(f"Loja {i}: Dados não disponíveis ou formato incorreto")
            aval.loc[f'Loja {i}', 'Média'] = None
            aval.loc[f'Loja {i}', 'Total de Avaliações'] = 0

    # Remover lojas sem média
    aval = aval.dropna(subset=['Média'])

    if not aval.empty:
        print("Média de Avaliação por Loja (ordem fixa):")
        display(aval.loc[['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']].style.format({
            'Média': '{:.2f}',
        }).background_gradient(cmap='Blues', subset=['Média']))
    else:
        print("Nenhum dado de avaliação válido encontrado.")

except Exception as e:
    print(f"Ocorreu um erro: {str(e)}")
    print("Verifique se as variáveis loja1, loja2, loja3 e loja4 são DataFrames válidos")
```

Produtos Mais e Menos Vendidos

```python
# Dicionário para armazenar os resultados
top_produtos = {}
bottom_produtos = {}

# Analise de cada loja
for i, df_loja in enumerate([loja1, loja2, loja3, loja4], 1):
    # Contagem de vendas por produto
    contagem_produtos = df_loja['Produto'].value_counts()

    # Produtos mais vendidos
    top_produtos[f'Loja {i}'] = contagem_produtos.head()

    # Produtos menos vendidos
    produtos_menos_vendidos = contagem_produtos[contagem_produtos > 0].tail() if len(contagem_produtos[contagem_produtos > 1]) >= 5 else contagem_produtos.tail(5)
    bottom_produtos[f'Loja {i}'] = produtos_menos_vendidos

# Resultados
print("=== PRODUTOS MAIS VENDIDOS POR LOJA ===")
for loja, produtos in top_produtos.items():
    print(f"\n{loja}:")
    display(produtos)

print("\n=== PRODUTOS MENOS VENDIDOS POR LOJA ===")
for loja, produtos in bottom_produtos.items():
    print(f"\n{loja}:")
    display(produtos)
```

Frete Médio por Loja

```python
# DataFrame para armazenar os resultados
frete_stats = pd.DataFrame()

# Calcular métricas para cada loja
for i, df_loja in enumerate([loja1, loja2, loja3, loja4], 1):
    if 'Frete' in df_loja.columns:
        df_loja['Frete'] = pd.to_numeric(df_loja['Frete'], errors='coerce')

        # Calcular estatísticas
        frete_stats.loc[f'Loja {i}', 'Frete Médio'] = df_loja['Frete'].mean()
    else:
        print(f"Aviso: Coluna 'Frete' não encontrada na Loja {i}")

# Ordenar por frete médio (do maior para o menor)
frete_stats = frete_stats.sort_values(by='Frete Médio', ascending=False)

# Exibir resultados
print("\nEstatísticas de Frete por Loja:")
display(frete_stats.style.format({
    'Frete Médio': 'R${:.2f}'
}).background_gradient(cmap='Reds', subset=['Frete Médio']))

## Resultados - Gráfico de Barras
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
frete_stats['Frete Médio'].plot(kind='bar', color='salmon', edgecolor='black', alpha=0.7)
plt.title('Frete Médio por Loja', fontsize=14)
plt.xlabel('Loja', fontsize=12)
plt.ylabel('Valor (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adicionando valores nas barras
for i, v in enumerate(frete_stats['Frete Médio']):
    plt.text(i, v + 0.5, f"R${v:.2f}", ha='center', fontsize=10)

## Análise Complementar
# Calcular frete como % do preço médio
for i, df_loja in enumerate([loja1, loja2, loja3, loja4], 1):
    if 'Frete' in df_loja.columns and 'Preço' in df_loja.columns:
        df_loja['Frete_%'] = (df_loja['Frete'] / df_loja['Preço']) * 100
        media_percentual = df_loja['Frete_%'].mean()
        print(f"Loja {i}: Frete representa em média {media_percentual:.1f}% do valor do produto")

print("\n=== CONCLUSÃO ===")
loja_maior_frete = frete_stats['Frete Médio'].idxmax()
loja_menor_frete = frete_stats['Frete Médio'].idxmin()
print(f"A {loja_maior_frete} tem o maior frete médio (R${frete_stats.loc[loja_maior_frete, 'Frete Médio']:.2f})")
print(f"A {loja_menor_frete} tem o menor frete médio (R${frete_stats.loc[loja_menor_frete, 'Frete Médio']:.2f})")
```

```python
# Atribuição por regiões
def assign_enhanced_region(row):
    lat, lon = row['lat'], row['lon']
    lat_median = np.median([loja1['lat'].median(), loja2['lat'].median(),
                           loja3['lat'].median(), loja4['lat'].median()])
    lon_median = np.median([loja1['lon'].median(), loja2['lon'].median(),
                           loja3['lon'].median(), loja4['lon'].median()])

    if lat > lat_median and lon > lon_median:
        return 'Nordeste (NE)'
    elif lat > lat_median and lon <= lon_median:
        return 'Noroeste (NW)'
    elif lat <= lat_median and lon > lon_median:
        return 'Sudeste (SE)'
    else:
        return 'Sudoeste (SW)'

for df in [loja1, loja2, loja3, loja4]:
    df['Região'] = df.apply(assign_enhanced_region, axis=1)

metrics = []
for i, df in enumerate([loja1, loja2, loja3, loja4], 1):
    region_stats = df.groupby('Região').agg({
        'Preço': ['sum', 'count'],
        'Avaliação da compra': 'mean',
        'Frete': 'mean',
        'lat': 'mean',
        'lon': 'mean'
    })
    region_stats.columns = ['Faturamento', 'Qtd Vendas', 'Avaliação Média', 'Frete Médio', 'Lat Média', 'Lon Média']
    region_stats['Loja'] = f'Loja {i}'
    region_stats['% Faturamento'] = 100 * region_stats['Faturamento'] / region_stats['Faturamento'].sum()
    metrics.append(region_stats)

regioes = pd.concat(metrics).reset_index()

## 2. Grid de Gráficos
plt.figure(figsize=(18, 12))
plt.suptitle('Análise de Desempenho por Região Geográfica', fontsize=18, y=1.02)

# Gráfico 1: Faturamento por região e loja
plt.subplot(2, 2, 1)
sns.barplot(x='Região', y='Faturamento', hue='Loja',
            data=regioes, palette='mako', edgecolor='black')
plt.title('Faturamento Total por Região', fontsize=14)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.xlabel('')
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Loja', bbox_to_anchor=(1.05, 1), loc='upper left')

# Gráfico 2: Distribuição percentual
plt.subplot(2, 2, 2)
sns.histplot(data=regioes, x='Região', weights='% Faturamento',
             hue='Loja', multiple='stack', palette='mako',
             shrink=0.8, edgecolor='black')
plt.title('Distribuição Percentual do Faturamento', fontsize=14)
plt.ylabel('Participação no Faturamento (%)', fontsize=12)
plt.xlabel('')
plt.grid(axis='y', alpha=0.3)

# Gráfico 3: Avaliação média
plt.subplot(2, 2, 3)
sns.pointplot(x='Região', y='Avaliação Média', hue='Loja',
              data=regioes, palette='mako', markers=['o', 's', 'D', '^'],
              linestyles=['-', '--', ':', '-.'])
plt.title('Satisfação do Cliente por Região', fontsize=14)
plt.ylabel('Avaliação Média (0-5)', fontsize=12)
plt.xlabel('')
plt.grid(axis='y', alpha=0.3)
plt.ylim(0, 5)

# Gráfico 4: Frete médio
plt.subplot(2, 2, 4)
sns.boxplot(x='Região', y='Frete Médio', hue='Loja',
            data=regioes, palette='mako', showfliers=False)
plt.title('Custo de Frete por Região', fontsize=14)
plt.ylabel('Frete Médio (R$)', fontsize=12)
plt.xlabel('')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

```

```python
def calcular_valores_normalizados(loja1, loja2, loja3, loja4):
    # 1. Faturamento
    faturamento_total = {
        'Loja 1': loja1['Preço'].sum(),
        'Loja 2': loja2['Preço'].sum(),
        'Loja 3': loja3['Preço'].sum(),
        'Loja 4': loja4['Preço'].sum()
    }
    max_fat = max(faturamento_total.values())
    fat_norm = {k: v/max_fat for k, v in faturamento_total.items()}

    # 2. Avaliação
    aval_norm = {
        'Loja 1': loja1['Avaliação da compra'].mean() / 5,
        'Loja 2': loja2['Avaliação da compra'].mean() / 5,
        'Loja 3': loja3['Avaliação da compra'].mean() / 5,
        'Loja 4': loja4['Avaliação da compra'].mean() / 5
    }

    # 3. Frete
    frete_medio = {
        'Loja 1': loja1['Frete'].mean(),
        'Loja 2': loja2['Frete'].mean(),
        'Loja 3': loja3['Frete'].mean(),
        'Loja 4': loja4['Frete'].mean()
    }
    max_frete = max(frete_medio.values())
    frete_norm = {k: 1 - (v/max_frete) for k, v in frete_medio.items()}

    # 4. Categoria
    diversidade = {
        'Loja 1': len(loja1['Categoria do Produto'].unique()),
        'Loja 2': len(loja2['Categoria do Produto'].unique()),
        'Loja 3': len(loja3['Categoria do Produto'].unique()),
        'Loja 4': len(loja4['Categoria do Produto'].unique())
    }
    div_norm = {k: v/10 for k, v in diversidade.items()}

    # Combinar todos os valores
    valores_finais = {
        'Loja 1': [fat_norm['Loja 1'], aval_norm['Loja 1'], frete_norm['Loja 1'], div_norm['Loja 1']],
        'Loja 2': [fat_norm['Loja 2'], aval_norm['Loja 2'], frete_norm['Loja 2'], div_norm['Loja 2']],
        'Loja 3': [fat_norm['Loja 3'], aval_norm['Loja 3'], frete_norm['Loja 3'], div_norm['Loja 3']],
        'Loja 4': [fat_norm['Loja 4'], aval_norm['Loja 4'], frete_norm['Loja 4'], div_norm['Loja 4']]
    }

    return valores_finais

valores_reais = calcular_valores_normalizados(loja1, loja2, loja3, loja4)
print(valores_reais)
```

```python
#Gráfico de Barras Empilhadas (Composição de Vendas por Categoria)

def plot_vendas_por_categoria(loja1, loja2, loja3, loja4):
    try:

        categorias_lojas = pd.DataFrame()

        for i, df_loja in enumerate([loja1, loja2, loja3, loja4], 1):
            # Verificar se a coluna existe e converter contagem para porcentagem
            if 'Categoria do Produto' in df_loja.columns:
                contagem = df_loja['Categoria do Produto'].value_counts(normalize=True) * 100
                categorias_lojas[f'Loja {i}'] = contagem
            else:
                print(f"Coluna 'Categoria do Produto' não encontrada na Loja {i}")
                return

        # Preencher NaN com 0 e ordenar
        categorias_lojas = categorias_lojas.fillna(0)

        # Ordenanado pela soma total
        categorias_lojas['Total'] = categorias_lojas.sum(axis=1)
        categorias_lojas = categorias_lojas.sort_values(by='Total', ascending=False)
        categorias_lojas = categorias_lojas.drop(columns='Total')

        # Gráfico
        plt.figure(figsize=(14, 8))
        ax = categorias_lojas.plot(kind='bar', stacked=True,
                                 colormap='tab20',
                                 edgecolor='black',
                                 width=0.8,
                                 figsize=(14, 8))

        plt.title('Distribuição Percentual de Vendas por Categoria', fontsize=16, pad=20)
        plt.xlabel('Categoria de Produto', fontsize=12)
        plt.ylabel('Percentual de Vendas (%)', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.legend(title='Loja', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Rótulos totais
        for x in range(len(categorias_lojas)):
            # Posição acumulada para cada categoria
            cum_sum = 0
            for i, loja in enumerate(categorias_lojas.columns):
                value = categorias_lojas.iloc[x, i]
                if value > 0:
                    ax.text(x, cum_sum + value/2, f"{value:.1f}%",
                           ha='center', va='center',
                           fontsize=8, color='white')
                cum_sum += value

        plt.tight_layout()
        plt.show()

        return categorias_lojas

    except Exception as e:
        print(f"Ocorreu um erro: {str(e)}")
        print("Verifique se todas as lojas têm a coluna 'Categoria do Produto'")
        return None

resultado = plot_vendas_por_categoria(loja1, loja2, loja3, loja4)
```
